# LLM as a Judge
Use an LLM to judge AI agent output. These judgements will help improve the system's output when given bad inputs.

```python
from deepeval.metrics import GEval
from deepeval.test_case import LLMTestCase, LLMTestCaseParams
from deepeval import evaluate
```

## Define Custom LLM-as-a-Judge Metric

```python
# Create a custom G-Eval metric for evaluating LLM outputs
correctness_metric = GEval(
    name="Correctness",
    criteria="Determine whether the actual output is factually correct based on the expected output.",
    evaluation_params=[
        LLMTestCaseParams.INPUT,
        LLMTestCaseParams.ACTUAL_OUTPUT,
        LLMTestCaseParams.EXPECTED_OUTPUT,
    ],
)
```

## Create Test Cases

```python
# Define test cases with input, actual output, and expected output
test_cases = [
    LLMTestCase(
        input="What is the capital of France?",
        actual_output="The capital of France is Paris.",
        expected_output="Paris",
    ),
    LLMTestCase(
        input="What is 2 + 2?",
        actual_output="2 + 2 equals 5.",
        expected_output="4",
    ),
    LLMTestCase(
        input="Explain photosynthesis.",
        actual_output="Photosynthesis is the process by which plants convert sunlight into energy.",
        expected_output="Plants use sunlight, water, and CO2 to produce glucose and oxygen.",
    ),
]
```

## Run Evaluation

```python
# Run evaluation with the custom metric
evaluate(test_cases=test_cases, metrics=[correctness_metric])
```

```python
# Or run individual metric measurement
for test_case in test_cases:
    correctness_metric.measure(test_case)
    print(f"Input: {test_case.input}")
    print(f"Score: {correctness_metric.score}")
    print(f"Reason: {correctness_metric.reason}\n")
```
